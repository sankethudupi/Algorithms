### Dealing with Null Values

```python
df0 = df.fillna(0)
df0.sample(5)
```

### Getting the distinct items in dataset


```python
distinct_items = {"whole milk"}
for index, row in df0.iterrows(): 
    #print (row["Name"], row["Age"])
    for i in range(1,33):
        if(row["Item " + str(i)]):
            distinct_items.add(row["Item " + str(i)])
print(distinct_items)
```

### Converting our DATA SET into Binary Transaction Format

```python
tb_data = pd.DataFrame(columns = distinct_items)
tb_data.reset_index()
```

```python

tb_data.loc[0,"sugar"] = 1
```

```python
tb_data
```

```python
for index, row in df0.iterrows(): 
    #print (row["Name"], row["Age"])
    for i in range(1,33):
        if(row["Item " + str(i)]) in distinct_items:
            tb_data.loc[index,row["Item " + str(i)]] = 1
```

```python

tb_data = tb_data.fillna(0)
```

### Importing Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

### Since it is a large Data Set with 9800 Transaction , min_support of 0.01 is chosen

```python

freq_itemset = apriori(df = tb_data, min_support = 0.005, use_colnames = True,)
freq_itemset.sort_values(by = 'support',ascending = False)
```

```python
#Forming Association Rules with a confidence threshold of 0.1
```

```python
rules = association_rules(freq_itemset, min_threshold = 0.1,metric="confidence") # by default, theshold metric is conf
#rules['consequents'] = rules['consequents'].apply(lambda x: list(x)[0]).astype("unicode")
#rules['antecedents'] = rules['antecedents'].apply(lambda x: list(x)).astype("unicode")
rules.head(5)
```

```python
type(rules.consequents)
```

## Whole milk and other vegetables are very common in all rules. so exempting them from consequents to find other important associations

```python
s1 = frozenset({'whole milk'})
s2 = frozenset({'other vegetables'})
x = pd.DataFrame()
for index, row in rules.iterrows():
    if (row["consequents"].isdisjoint(s1) & row["consequents"].isdisjoint(s2)):
        x = x.append(pd.DataFrame([list(row)], columns =rules.columns))
x.reset_index(drop=True, inplace=True)
x.sort_values(by = ["confidence","lift"], ascending = False)
```
